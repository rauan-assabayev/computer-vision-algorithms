```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
import random
```

```python
image = data.camera()
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        image[i,j] += random.randint(-20,20)
```

```python
plt.imshow(image, cmap = plt.cm.gray)
```

```python
mm = np.median(image)
print(mm)
```

```python
binary = image > mm
```

```python
plt.imshow(binary, cmap = plt.cm.gray)
```

```python
from skimage import feature

edges1 = feature.canny(image,sigma=2,)
```

```python
plt.imshow(edges1, cmap = plt.cm.gray)
```

```python
edges2 = feature.canny(image,sigma=4.5,low_threshold=10.0,high_threshold=8.0)
plt.imshow(edges2, cmap = plt.cm.gray)
plt.title("Noize")
```

```python
image = data.camera()
binary = image > mm
edges0 = feature.canny(image,sigma=2,)
plt.imshow(edges0, cmap = plt.cm.gray)
plt.title("Original")
```
